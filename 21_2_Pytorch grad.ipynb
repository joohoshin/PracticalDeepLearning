2w**2 + 5 에 대해 미분해보자


```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.linspace(1.5,2.5,num=1000)
```

```python
z = 2*w**w + 5
```

```python
plt.plot(w, z)
```

```python
w = torch.tensor(2.0, requires_grad=True)
```

```python
z=2*w**2 + 5
```

```python
z
```

```python
z.backward()   #자동 미분 계산
```

```python
w.grad
```
