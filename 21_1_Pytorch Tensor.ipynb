```python
import numpy as np
import torch
```

```python
# 리스트로부터 텐서를 만들수 있다.

t1 = torch.tensor([[1,2],[3,4]])
```

```python
# 타입 지정 가능

t2 = torch.tensor([[1,2],[3,4]], dtype=torch.float64)
```

pytorch는 device에 대해 신경써줘야 함


```python
t1.device   #gpu 런타임 실행이나, 텐서는 cpu에 타입으로 저장
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

if device.type == 'cuda':
    print(torch.cuda.get_device_name(0))
```

```python
device
```

```python
t3 = torch.tensor([[1,2],[3,4]], device = 'cuda')  #device 지정
```

```python
t3.device
```

```python
t4 = torch.tensor([[1,2],[3,4]], device = 'cuda:0')
```

```python
t4.device
```

Numpy로 변환 시 gpu device의 tensor는 cpu로 이동 후 변환

```python
t4.numpy()  # gpu device로 설정된 텐서를 numpy로 변환시 에러가 발생
```

```python
t4.to('cpu').numpy()  # cpu 변환 후 numpy로 변환
```

tensor의 인덱싱 및 연산은 numpy와 거의 동일

```python
t4[:,1]
```

```python
t3 + t4
```

```python
t1+t4   # t1이 cpu인 경우 gpu의 t4와 연산 안됨, 동일한 device로 맞춰줘야 함
```

```python
t1+t4.to('cpu')
```

데이터 병합 시 torch.cat() 사용

numpy의 reshape은 view() 사용

```python
t5=torch.tensor([[1,2],[3,4]])
t6=torch.tensor([[10,20,30],[30,40,50]])
```

```python
t5
```

```python
t6
```

```python
t6.size()
```

```python
t6.t()
```

```python
torch.cat([t5,t6], dim=1)
```

```python
t7=torch.cat([t5,t6.t()], dim=0)
```

```python
t7
```

```python
t7.view(1,-1)  #-1은 자동 개수
```

```python
t7.view(2,5)
```
